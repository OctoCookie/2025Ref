```python


# функція, яка читає, чистить, додає RUL і малює важливості
def process_and_plot(path_in, path_out_csv):
    # 1) Задаємо імена стовпців
    col_names = (
        ["engineNumber", "cycleNumber"] +
        [f"setting{i}" for i in range(1, 4)] +
        [f"sensor{i}"     for i in range(1, 22)]
    )

    # 2) Читаємо
    df = pd.read_csv(path_in, sep=r"\s+", header=None, names=col_names)

    # 3) Видаляємо непотрібні сенсори
    cols_to_remove = ['sensor1','sensor5','sensor10','sensor19','sensor18','sensor6', 'sensor16']
    existing = [c for c in cols_to_remove if c in df.columns]
    df.drop(existing, axis=1, inplace=True)

    # 4) Обчислюємо RUL
    df['RUL'] = df.groupby('engineNumber')['cycleNumber'].transform('max') - df['cycleNumber']

    # 5) Формуємо X та y
    features = [c for c in df.columns if c not in ('engineNumber','cycleNumber','RUL')]
    X, y = df[features], df['RUL']

    # 6) Тренуємо модель і малюємо важливості
    model = RandomForestRegressor(random_state=42).fit(X, y)
    imp = pd.Series(model.feature_importances_, index=features).sort_values()
    plt.figure(figsize=(8,6))
    imp.plot(kind='barh')
    plt.title(path_in)
    plt.show()

    # 7) Зберігаємо готовий CSV
    df.to_csv(path_out_csv, index=False)


# 8) Перелік файлів
for prefix in ('train','test'):
    for idx in range(1,5):
        inp  = f"C:/Users/Admin/111/kaggle/{prefix}_FD00{idx}.txt"
        outp = f"C:/Users/Admin/111/kaggle/{prefix}_FD00{idx}_processed.csv"
        process_and_plot(inp, outp)
```

```python
# ————— 1. Зчитуємо дані —————
train = pd.read_csv(r"C:/Users/Admin/111/kaggle/train_all_processed.csv")
test  = pd.read_csv(r"C:/Users/Admin/111/kaggle/test_last_with_RUL.csv")

# ————— 2. Відбираємо ознаки та ціль —————
# Припускаємо, що у таблицях є стовпець 'RUL' ідентичний у обох
feature_cols = [c for c in train.columns 
                if c not in ['engineNumber','cycleNumber','dataset','RUL']]

X_train = train[feature_cols]
y_train = train['RUL']

X_test  = test[ feature_cols ]
y_test  = test['RUL']

# ————— 3. RandomForestRegressor —————
rf = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)

# Метрики
print("=== RandomForest ===")
print("MAE:",  mean_absolute_error(y_test, rf_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, rf_pred)))
print("R2:  ", r2_score(y_test, rf_pred))

# ————— 4. XGBoostRegressor —————
xgb = XGBRegressor(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=5,
    random_state=42,
    n_jobs=-1,
    verbosity=0
)
xgb.fit(X_train, y_train)
xgb_pred = xgb.predict(X_test)

print("\n=== XGBoost ===")
print("MAE:",  mean_absolute_error(y_test, xgb_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, xgb_pred)))
print("R2:  ", r2_score(y_test, xgb_pred))

# ————— 5. (Опційно) Порівняльний DataFrame —————
results = pd.DataFrame({
    'y_true': y_test,
    'RF_pred': rf_pred,
    'XGB_pred': xgb_pred
}).reset_index(drop=True)

print("\nПара прикладів прогнозів:")
print(results.head(10))
```

```python
# 1) Зчитуємо датасети
train = pd.read_csv(r"C:/Users/Admin/111/kaggle/train_all_processed.csv")
test  = pd.read_csv(r"C:/Users/Admin/111/kaggle/test_last_with_RUL.csv")

# 2) Обираємо ознаки і ціль
features = [c for c in train.columns 
            if c not in ['engineNumber','cycleNumber','dataset','RUL']]
X_train, y_train = train[features], train['RUL']
X_test,  y_test  = test[features],  test['RUL']

# 3) Навчаємо моделі
rf  = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1).fit(X_train, y_train)
xgb = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=5,
                   random_state=42, n_jobs=-1, verbosity=0).fit(X_train, y_train)

# 4) Робимо прогнози
test['RF_pred']  = rf.predict(X_test)
test['XGB_pred'] = xgb.predict(X_test)


# 5) Оцінюємо калібровку: лінійна регресія y_true ~ y_pred
def calibration_stats(y_true, y_pred, name):
    lr = LinearRegression().fit(y_pred.reshape(-1,1), y_true)
    m, b = lr.coef_[0], lr.intercept_
    print(f"{name} calibration: slope = {m:.3f}, intercept = {b:.3f}")
    return m, b

m_rf, b_rf   = calibration_stats(y_test.values, test['RF_pred'].values,  'RandomForest')
m_xgb, b_xgb = calibration_stats(y_test.values, test['XGB_pred'].values, 'XGBoost')

# 7) Можна також накласти трендові лінії на повний scatter
plt.figure(figsize=(6,6))
plt.scatter(y_test, test['RF_pred'], s=20, alpha=0.3, label='RF scatter',  c='C0')
plt.scatter(y_test, test['XGB_pred'], s=20, alpha=0.3, label='XGB scatter', c='C1')
# Ідеальна діагональ
lims = [0, max(y_test.max(), test['RF_pred'].max(), test['XGB_pred'].max())]
plt.plot(lims, lims, '--', color='gray')
# Трендові лінії
plt.plot(lims, [m_rf*lim + b_rf for lim in lims], '-', c='C0', label='RF calib line')
plt.plot(lims, [m_xgb*lim + b_xgb for lim in lims], '-', c='C1', label='XGB calib line')

plt.xlabel("True RUL")
plt.ylabel("Predicted RUL")
plt.title("Calibration plot")
plt.legend()
plt.tight_layout()
plt.show()
```

```python


# 1) Зчитуємо дані
train = pd.read_csv(r"C:/Users/Admin/111/kaggle/train_all_processed.csv")
test  = pd.read_csv(r"C:/Users/Admin/111/kaggle/test_last_with_RUL.csv")

features = [c for c in train.columns 
            if c not in ['engineNumber','cycleNumber','dataset','RUL']]
X_train, y_train = train[features], train['RUL']
X_test,  y_test  = test[features],  test['RUL']

# 2) Навчаємо обидві моделі
rf  = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1).fit(X_train, y_train)
xgb = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=5,
                   random_state=42, n_jobs=-1, verbosity=0).fit(X_train, y_train)

# 3) Базові прогнози
rf_pred  = rf.predict(X_test)
xgb_pred = xgb.predict(X_test)

# 4) Ваші калібрувальні коефіцієнти
m_rf, b_rf   = 0.677, 13.376
m_xgb, b_xgb = 0.679, 13.522

# 5) Калібруємо прогнози
rf_cal = (rf_pred  - b_rf)  / m_rf
xgb_cal= (xgb_pred - b_xgb) / m_xgb

# 6) Метрики до та після калібрування
def print_metrics(y_true, y_pred, label):
    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2   = r2_score(y_true, y_pred)
    print(f"{label}: MAE={mae:.2f}, RMSE={rmse:.2f}, R2={r2:.3f}")

print("=== Before calibration ===")
print_metrics(y_test, rf_pred,  "RF")
print_metrics(y_test, xgb_pred, "XGB")

print("\n=== After calibration ===")
print_metrics(y_test, rf_cal,  "RF_calibrated")
print_metrics(y_test, xgb_cal, "XGB_calibrated")

# 7) Графік: True vs Pred (до і після) для RF
plt.figure(figsize=(6,6))
plt.scatter(y_test, rf_pred,   s=12, alpha=0.3, label='RF raw',  color='C0')
plt.scatter(y_test, rf_cal,    s=12, alpha=0.3, label='RF calib', color='C1')
lims = [0, y_test.max()]
plt.plot(lims, lims, '--', color='gray')
plt.xlabel("True RUL")
plt.ylabel("Predicted RUL")
plt.title("RF: Before vs After Calibration")
plt.legend()
plt.tight_layout()
plt.show()

# 8) І те саме для XGB
plt.figure(figsize=(6,6))
plt.scatter(y_test, xgb_pred,   s=12, alpha=0.3, label='XGB raw',  color='C0')
plt.scatter(y_test, xgb_cal,    s=12, alpha=0.3, label='XGB calib', color='C1')
plt.plot(lims, lims, '--', color='gray')
plt.xlabel("True RUL")
plt.ylabel("Predicted RUL")
plt.title("XGB: Before vs After Calibration")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
def sliding_windows(sig: np.ndarray,
                    win_size: int = 1000,
                    step:     int = 500):
    """
    Розбиває сигнал sig (n_samples, n_channels)
    на послідовність вікон розміру win_size з кроком step.
    Повертає масив форми (n_windows, win_size, n_channels).
    """
    n, c = sig.shape
    starts = np.arange(0, n - win_size + 1, step)
    windows = [sig[i:i+win_size] for i in starts]
    return np.stack(windows, axis=0)
```

```python
from scipy.stats import skew, kurtosis

def extract_time_features(win: np.ndarray):
    """
    win: (win_size, n_channels)
    Повертає dict із ознаками для всіх каналів.
    """
    feats = {}
    for ch in range(win.shape[1]):
        x = win[:, ch]
        key = f"ch{ch+1}"
        feats[f"{key}_mean"]   = np.mean(x)
        feats[f"{key}_std"]    = np.std(x)
        feats[f"{key}_rms"]    = np.sqrt(np.mean(x**2))
        feats[f"{key}_skew"]   = skew(x)
        feats[f"{key}_kurt"]   = kurtosis(x)
        feats[f"{key}_crest"]  = np.max(np.abs(x)) / (feats[f"{key}_rms"] + 1e-8)
    return feats
```

```python
from scipy.fft import rfft, rfftfreq

def extract_freq_features(win: np.ndarray, fs: int = 1000):
    feats = {}
    n = win.shape[0]
    freqs = rfftfreq(n, 1/fs)
    for ch in range(win.shape[1]):
        x = win[:, ch]
        X = np.abs(rfft(x))
        # спектральний центр
        sc = np.sum(freqs * X) / (np.sum(X)+1e-8)
        # спектральна ширина
        sw = np.sqrt(np.sum(((freqs - sc)**2) * X) / (np.sum(X)+1e-8))
        feats[f"ch{ch+1}_spec_cent"]  = sc
        feats[f"ch{ch+1}_spec_bw"]    = sw
        # енергія в трьох бендах
        for (low, hi, name) in [(5,50,'lo'), (50,200,'mid'), (200,500,'hi')]:
            mask = (freqs>=low)&(freqs<hi)
            feats[f"ch{ch+1}_en_{name}"] = np.sum(X[mask]**2)
    return feats
```

```python
import pandas as pd

all_features = []
all_labels   = []  # 0 або 1 для Healthy/Faulty

# пробігаємо по всіх файлах
for label, folder in [(0, healthy_dir), (1, faulty_dir)]:
    for path in glob.glob(os.path.join(folder, "*.mat")):
        sig = loadmat(path)[next(k for k in loadmat(path) if not k.startswith("__"))][:, :3]
        # фільтруємо (як ви вже зробили) кожний канал перед цією частиною
        # sig = filtfilt(b,a,sig[:,0]) etc...
        
        windows = sliding_windows(sig, win_size=1000, step=500)
        for win in windows:
            feats = {}
            feats.update(extract_time_features(win))
            feats.update(extract_freq_features(win, fs=1000))
            feats["file"] = os.path.basename(path)
            all_features.append(feats)
            all_labels.append(label)


df_feats = pd.DataFrame(all_features)
df_feats['label'] = all_labels

print("Features shape:", df_feats.shape)
print(df_feats.head())
df_feats.to_csv("df_feats.csv", index=False)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import (
    GroupShuffleSplit,
    RandomizedSearchCV
)
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, classification_report

# 0) Зчитаємо всі ознаки з файлу (має бути з колонками file + label + числові фічі)
df_feats = pd.read_csv("df_feats.csv")

# 1) Виділимо список числових фіч (всі колонки крім file та label)
feature_cols = [c for c in df_feats.columns if c not in ("file","label")]

# 2) Масштабуємо ці фічі через StandardScaler
scaler = StandardScaler()
df_feats_scaled = df_feats.copy()
df_feats_scaled[feature_cols] = scaler.fit_transform(df_feats[feature_cols])

# 3) Відокремимо X, y та groups для спліту
X      = df_feats_scaled[feature_cols]
y      = df_feats_scaled["label"]
groups = df_feats_scaled["file"]

# 4) Group-wise split за файлом (20% файлів у тест)
gss = GroupShuffleSplit(n_splits=1, test_size=0.20, random_state=42)
train_idx, test_idx = next(gss.split(X, y, groups=groups))

X_tr, X_te = X.iloc[train_idx], X.iloc[test_idx]
y_tr, y_te = y.iloc[train_idx], y.iloc[test_idx]

print("TRAIN files:", np.unique(groups.iloc[train_idx]))
print(" TEST files:", np.unique(groups.iloc[test_idx]))

# 5) Базове навчання RandomForest
rf = RandomForestClassifier(
    n_estimators=100,
    n_jobs=-1,          # паралельно будує дерева
    random_state=42
)
rf.fit(X_tr, y_tr)
rf_proba = rf.predict_proba(X_te)[:,1]
rf_pred  = (rf_proba >= 0.5).astype(int)

print("\n=== Base RandomForest ===")
print("ROC AUC:", roc_auc_score(y_te, rf_proba))
print(classification_report(y_te, rf_pred, target_names=["Healthy","Faulty"]))

# 6) Базове навчання XGBoost
xgb = XGBClassifier(
    eval_metric="logloss",
    n_jobs=-1,
    random_state=42
)
xgb.fit(X_tr, y_tr)
xgb_proba = xgb.predict_proba(X_te)[:,1]
xgb_pred  = (xgb_proba >= 0.5).astype(int)

print("=== Base XGBoost ===")
print("ROC AUC:", roc_auc_score(y_te, xgb_proba))
print(classification_report(y_te, xgb_pred, target_names=["Healthy","Faulty"]))

# 7) RandomizedSearchCV для RandomForest — лише CV паралельно
param_dist = {
    "n_estimators":     [100, 200, 500],
    "max_depth":        [None, 5, 10, 20],
    "min_samples_leaf": [1, 2, 5],
    "max_features":     ["sqrt", "log2", 0.5],
}

rs = RandomizedSearchCV(
    RandomForestClassifier(
        random_state=42,
        n_jobs=1         # RF однопоточний всередині кожного фолду
    ),
    param_distributions=param_dist,
    n_iter=20,
    cv=3,
    scoring="roc_auc",
    n_jobs=1,           # лише ОДИН потік тут, щоб уникнути TerminatedWorkerError
    random_state=42,
    verbose=1
)
rs.fit(X_tr, y_tr)

print("\nBest RF params:", rs.best_params_)
best_rf       = rs.best_estimator_
best_rf_proba = best_rf.predict_proba(X_te)[:,1]
best_rf_pred  = (best_rf_proba >= 0.5).astype(int)

print("=== Tuned RandomForest ===")
print("ROC AUC:", roc_auc_score(y_te, best_rf_proba))
print(classification_report(y_te, best_rf_pred, target_names=["Healthy","Faulty"]))
```

```python
from sklearn.isotonic import IsotonicRegression
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import roc_auc_score, brier_score_loss
import matplotlib.pyplot as plt
from sklearn.calibration import calibration_curve

# 1) Зробимо “raw” ймовірності на тесті
raw_rf_te  = best_rf.predict_proba(X_te)[:, 1]
raw_xgb_te = xgb.predict_proba(X_te)[:, 1]

# 2) OOF-прогнози на train для калібрування
oof_rf  = cross_val_predict(best_rf, X_tr, y_tr,
                            cv=5,
                            method="predict_proba",
                            n_jobs=1)[:, 1]

oof_xgb = cross_val_predict(xgb, X_tr, y_tr,
                            cv=5,
                            method="predict_proba",
                            n_jobs=1)[:, 1]

# 3) Навчаємо ізотоніки
iso_rf  = IsotonicRegression(out_of_bounds="clip").fit(oof_rf,  y_tr)
iso_xgb = IsotonicRegression(out_of_bounds="clip").fit(oof_xgb, y_tr)

# 4) Калібруємо тестові ймовірності
calib_rf_te  = iso_rf.transform(raw_rf_te)
calib_xgb_te = iso_xgb.transform(raw_xgb_te)

# 5) Метрики до/після калібрування
print("=== RandomForest ===")
print(f"ROC-AUC raw: {roc_auc_score(y_te, raw_rf_te):.3f}")
print(f"ROC-AUC iso: {roc_auc_score(y_te, calib_rf_te):.3f}")
print(f"Brier  raw: {brier_score_loss(y_te, raw_rf_te):.4f}")
print(f"Brier  iso: {brier_score_loss(y_te, calib_rf_te):.4f}\n")

print("=== XGBoost ===")
print(f"ROC-AUC raw: {roc_auc_score(y_te, raw_xgb_te):.3f}")
print(f"ROC-AUC iso: {roc_auc_score(y_te, calib_xgb_te):.3f}")
print(f"Brier  raw: {brier_score_loss(y_te, raw_xgb_te):.4f}")
print(f"Brier  iso: {brier_score_loss(y_te, calib_xgb_te):.4f}")

# 6) (Опційно) Reliability diagram
plt.figure(figsize=(6,5))
for preds, name in [(raw_rf_te, "RF raw"),
                    (calib_rf_te, "RF iso"),
                    (raw_xgb_te, "XGB raw"),
                    (calib_xgb_te, "XGB iso")]:
    frac_pos, mean_pred = calibration_curve(y_te, preds, n_bins=10)
    plt.plot(mean_pred, frac_pos, marker='o', label=name)
plt.plot([0,1],[0,1],'k--', label="perfect")
plt.xlabel("Predicted probability")
plt.ylabel("Fraction of positives")
plt.legend()
plt.title("Reliability diagram")
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve

# приклад для RF:
prec, rec, thresh = precision_recall_curve(y_te, calib_rf_te)
# наприклад, знайдемо поріг, де precision≈recall
idx = np.argmin(np.abs(prec-rec))
best_thresh = thresh[idx]
print("Best thresh (RF):", best_thresh, "→ P/R:", prec[idx], rec[idx])
```

```python
rf_final = (calib_rf_te >= best_thresh).astype(int)
print(classification_report(y_te, rf_final, target_names=["H","F"]))
```

```python
# 1) розрахунок precision–recall для XGB
prec_x, rec_x, thresh_x = precision_recall_curve(y_te, calib_xgb_te)
idx_x = np.argmin(np.abs(prec_x - rec_x))
best_thresh_x = thresh_x[idx_x]
print("Best thresh (XGB):", best_thresh_x,
      "→ P/R:", prec_x[idx_x], rec_x[idx_x])

# 2) фінальний прогноз
xgb_final = (calib_xgb_te >= best_thresh_x).astype(int)
print(classification_report(y_te, xgb_final, target_names=["H","F"]))
```

```python
plt.figure(figsize=(6,4))
plt.plot(rec, prec, label="RF")
plt.plot(rec_x, prec_x, label="XGB")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision–Recall curve")
plt.legend()
plt.show()
```
